###KPI Requirements:







```python
# Total Sales
total_sales = df['Sales'].sum()

# Avg sales
avg_sales = df['Sales'].mean()

# No of item sold
items = df['Sales'].count()

# Avg Rating
avg_rating = df['Rating'].mean()


# print('Total Sales:', total_sales)
# print('Avg Sales:', avg_sales)
# print('No of items sold:', items)
# print('Avg Rating:', avg_rating)

print(f'Total Sales: ${total_sales:.2f}')
print(f'Avg Sales: ${avg_sales:.2f}')
print(f'No of items sold: {items:.0f}')
print(f'Avg Rating: {avg_rating:.2f}')
```

##CHART REQ


####Total Sales by Fat Content

```python
sales_by_fat = df.groupby('Item Fat Content')['Sales'].sum()

# colors = ['#66b3ff', '#ff9999', '#99ff99', '#ffcc99', '#c2c2f0'][:len(sales_by_fat)]
plt.figure(figsize=(5, 5))
wedges, texts, autotexts = plt.pie(
    sales_by_fat,
    labels=sales_by_fat.index,
    autopct='%.1f%%',
    startangle=90,
    # colors=colors,
    wedgeprops=dict(width=0.4, edgecolor='white'),
    shadow=True  # 👈 Adds depth
)

for text in texts:
    text.set_fontsize(11)
for autotexts in autotexts:
    autotexts.set_fontsize(10)
    autotexts.set_color('black')

plt.title('Total Sales by Fat Content', fontsize=20, fontweight='bold')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

####Total Sales by Item Type

```python
sales_by_item = df.groupby('Item Type')['Sales'].sum()

plt.figure(figsize=(15, 6))
bars = plt.bar(
    sales_by_item.index,
    sales_by_item.values,
    color='#66b3ff',
    edgecolor='black'
)

# Add value labels on top
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.01 * yval,
             f'{yval:.1f}', ha='center', va='bottom', fontsize=10)

# Rotate x labels
plt.xticks(rotation=45, ha='right', fontsize=12)

# Title and labels
plt.title('Total Sales by Item Type', fontsize=20, fontweight='bold')
plt.xlabel('Item Type', fontsize=14)
plt.ylabel('Total Sales', fontsize=14)

plt.tight_layout()
plt.show()
```

```python
# top 10 sales by item type
top_10 = sales_by_item.sort_values(ascending=False).head(10)

plt.figure(figsize=(11, 7))
bars = plt.bar(
    top_10.index,
    top_10.values,
    color='#66b3ff',
    edgecolor='black'
)

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.01 * yval,
             f'{yval:.1f}', ha='center', va='bottom', fontsize=10)

plt.xticks(rotation=45, ha='right', fontsize=12)
plt.title('Top 10 Total Sales by Item Type', fontsize=20, fontweight='bold')
plt.xlabel('Item Type', fontsize=14)
plt.ylabel('Total Sales', fontsize=14)
plt.tight_layout()
plt.show()
```

####Fat content by Outlet

```python
grouped = df.groupby(['Outlet Identifier', 'Item Fat Content']).agg(
    Total_Sales=('Sales', 'sum'),
    Avg_Sales=('Sales', 'mean'),
    Number_of_Items=('Sales', 'count'), # Corrected to use 'Sales' count
    Avg_Rating=('Rating', 'mean')
).reset_index()

pivot_sales = grouped.pivot(index='Outlet Identifier', columns='Item Fat Content', values='Total_Sales').fillna(0)

# Plot
ax = pivot_sales.plot(
    kind='bar',
    stacked=True,
    figsize=(10, 6),
    colormap='tab20'  # colorful palette
)

# Add data labels on each segment
for container in ax.containers:
    for bar in container:
        height = bar.get_height()
        if height > 0:  # Only label bars with height > 0
            ax.text(
                bar.get_x() + bar.get_width() / 2,
                bar.get_y() + height / 2,
                f'{height:,.0f}',  # Format with commas, no decimals
                ha='center',
                va='center',
                fontsize=9,
                color='black',
                fontweight='bold'
            )

plt.title('Total Sales by Fat Content and Outlet', fontsize=16, fontweight='bold')
plt.xlabel('Outlet', fontsize=14)
plt.ylabel('Total Sales', fontsize=14)
plt.xticks(rotation=0)
plt.legend(title='Fat Content', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

####Total Sales by Outlet Establishment

```python
# Calculate Outlet Age
current_year = pd.to_datetime('now').year
df['Outlet Age'] = current_year - df['Outlet Establishment Year']

# Group by Outlet Age (numeric, so line chart makes sense)
sales_by_age = df.groupby('Outlet Age')['Sales'].sum().sort_index()

plt.figure(figsize=(10, 6))
plt.plot(sales_by_age.index, sales_by_age.values, marker='o', linestyle='-', color='teal')

plt.title('Total Sales by Outlet Age', fontsize=16, fontweight='bold')
plt.xlabel('Outlet Age (years)', fontsize=14)
plt.ylabel('Total Sales', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(sales_by_age.index)  # show all age ticks; modify if too crowded
plt.tight_layout()
plt.show()
```

```python
# Group by Outlet Type and calculate total sales
sales_by_type = df.groupby('Outlet Type')['Sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
plt.plot(sales_by_type.index, sales_by_type.values, marker='o', linestyle='-', color='purple')

plt.title('Total Sales by Outlet Type', fontsize=16, fontweight='bold')
plt.xlabel('Outlet Type', fontsize=14)
plt.ylabel('Total Sales', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
sales_by_size = df.groupby('Outlet Size')['Sales'].sum()

# Define colors (optional)
colors = ['#66b3ff', '#ff9999', '#99ff99', '#ffcc99'][:len(sales_by_size)]

plt.figure(figsize=(6,6))
wedges, texts, autotexts = plt.pie(
    sales_by_size,
    labels=sales_by_size.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=colors,
    wedgeprops=dict(width=0.4, edgecolor='white')  # Donut effect
)

# Style text
for text in texts:
    text.set_fontsize(12)
for autotext in autotexts:
    autotext.set_fontsize(12)
    autotext.set_color('black')

plt.title('Total Sales by Outlet Size', fontsize=16, fontweight='bold')
plt.axis('equal')  # Equal aspect ratio ensures pie is circle
plt.tight_layout()
plt.show()
```
